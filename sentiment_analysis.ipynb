# Data Cleaning

```python
df = df.drop(['location'],axis=1)
```

```python
df = df.drop(['id'],axis = 1)
```

```python
df
```

```python
import string 
import re
import nltk
from nltk.stem import WordNetLemmatizer
import numpy as np
```

```python
cln = re.compile('<.*?>')
```

```python
def omit_tags(text):
  clntext = re.sub(cln, '', text)
  return clntext

def omit_punctuation(text):
  text = text.replace('"', "")
  text = text.replace("'", "")
  stop_char = string.punctuation
  no_punct="".join([i for i in text if i not in stop_char])
  return no_punct


df['text']= df['text'].apply(lambda x:omit_tags(x))
df['text']= df['text'].apply(lambda x:omit_punctuation(x))
```

```python
df
```

```python
def tokenizer(text):
  tokens = re.split('\W+',text)
  return tokens
```

```python
df['text'] = df['text'].apply(lambda x:tokenizer(x))
```

```python
df
```

```python
nltk.download('stopwords')
x = nltk.corpus.stopwords.words('english')
x[0:32]
```

```python
def ommiting_stopwords(a):
  y = [z for z in a if z not in x]
  return y
```

```python
df['text'] = df['text'].apply(lambda x: ommiting_stopwords(x))
```

```python
df
```

```python
lematizer = WordNetLemmatizer()
nltk.download('wordnet')
def lematize(x):
  lt = [lematizer.lemmatize(word) for word in x]
  return lt
```

```python
df['text']= df['text'].apply(lambda x:lematize(x))
```

```python
df
```

# LSTM Model

```python
# Importing required libraries
import nltk
import pandas as pd
from nltk.corpus import stopwords
from textblob import Word
from sklearn.preprocessing import LabelEncoder
from collections import Counter
import wordcloud
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
```

```python
# Encoded the target column
lb=LabelEncoder()
df['target'] = lb.fit_transform(df['target'])
```

```python
tokenizer = Tokenizer(num_words=500, split=' ') 
tokenizer.fit_on_texts(df['text'].values)
X = tokenizer.texts_to_sequences(df['text'].values)
X = pad_sequences(X)
```

```python
model = Sequential()
model.add(Embedding(500, 120, input_length = X.shape[1]))
model.add(SpatialDropout1D(0.4))
model.add(LSTM(176, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
print(model.summary())
```

```python
y=pd.get_dummies(df['target'])
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
import tensorflow as tf
```

```python
acc_checkpoint= tf.keras.callbacks.ModelCheckpoint(
    ('/content/drive/MyDrive/Best /sentiment_analysis.h5'),
    monitor="val_loss",
    verbose=0,
    save_best_only=False,
    save_weights_only=False,
    mode="auto",
    save_freq="epoch",
    options=None
)
```

```python
batch_size=32
model.fit(X_train, y_train, epochs = 80, callbacks=[acc_checkpoint] ,batch_size=batch_size, verbose = 'auto')
```
